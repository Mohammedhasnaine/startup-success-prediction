```python
#Cell 1 - Importing all libraries
# ============================================================
# MODULE 1: DATA PREPROCESSING & CLEANING
# Project: Startup Success Prediction
# ============================================================

# Core data libraries
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Preprocessing tools
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer

# For handling class imbalance later
from imblearn.over_sampling import SMOTE

# System & file operations
import os
import warnings
warnings.filterwarnings('ignore')

# Set plot style globally
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('husl')

print(" All libraries imported successfully!")
```

```python
# Cell 3 — load dataset and first look
# ============================================================
# STEP 1: LOAD THE DATASET
# ============================================================

df = pd.read_csv(DATASET_PATH)

print("=" * 55)
print("DATASET OVERVIEW")
print("=" * 55)
print(f"Rows    : {df.shape[0]}")
print(f"Columns : {df.shape[1]}")
print(f"\nColumn names:")
for i, col in enumerate(df.columns, 1):
    print(f"  {i:2}. {col}")

# Show first 5 rows
print("\nFirst 5 rows:")
df.head()
```

```python
# Cell 4 — check data types and missing values
# ============================================================
# STEP 2: INSPECT DATA QUALITY
# ============================================================

print("DATA TYPES:")
print(df.dtypes)

print("\nMISSING VALUES (columns with any missing):")
missing = df.isnull().sum()
missing = missing[missing > 0]
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
print(missing_df)

print(f"\nTarget column value counts:")
print(df['status'].value_counts())
print(f"\nClass balance: {df['status'].value_counts(normalize=True).round(3) * 100}")
```

```python
#Cell 5 — statistical summary
# ============================================================
# STEP 3: STATISTICAL SUMMARY
# ============================================================

print("STATISTICAL SUMMARY (numeric columns):")
df.describe().T.style.background_gradient(cmap='Blues')
```

```python
#Cell 6 — drop useless columns
# ============================================================
# STEP 4: DROP JUNK / USELESS COLUMNS
# These columns are IDs, duplicates, or >50% missing
# ============================================================

columns_to_drop = [
    'Unnamed: 0',      # Just row numbers — meaningless
    'Unnamed: 6',      # 53% missing, no useful info
    'id',              # Unique ID — not a feature
    'object_id',       # Duplicate unique ID
    'name',            # Company name — can't use text in ML
    'labels',          # Duplicate of target column
    'state_code.1',    # Duplicate of state_code
    'zip_code',        # Too granular, not predictive
    'closed_at',       # 63% missing + causes data leakage
    'founded_at',      # Date string — we extract features from it below
    'first_funding_at',# Date string — already captured in age columns
    'last_funding_at', # Date string — already captured in age columns
]

df.drop(columns=columns_to_drop, inplace=True)

print(f"Dropped {len(columns_to_drop)} columns")
print(f"Remaining columns: {df.shape[1]}")
print(f"Remaining columns list: {list(df.columns)}")
```

```python
#Cell 7 — handle missing values intelligently
# ============================================================
# STEP 5: HANDLE MISSING VALUES
# ============================================================

# Check missing values after dropping columns
print("Missing values remaining:")
print(df.isnull().sum()[df.isnull().sum() > 0])

# age_first_milestone_year and age_last_milestone_year
# have ~152 missing each. Fill with MEDIAN (not mean)
# because median is not affected by outliers

milestone_cols = ['age_first_milestone_year', 'age_last_milestone_year']
for col in milestone_cols:
    median_val = df[col].median()
    df[col].fillna(median_val, inplace=True)
    print(f" Filled '{col}' with median = {median_val:.2f}")

# city and state_code — fill with mode (most common value)
for col in ['city', 'state_code', 'category_code']:
    if col in df.columns and df[col].isnull().sum() > 0:
        mode_val = df[col].mode()[0]
        df[col].fillna(mode_val, inplace=True)
        print(f" Filled '{col}' with mode = {mode_val}")
print(f"\n Missing values after treatment: {df.isnull().sum().sum()}")
```

```python
# Cell 8 — encode target column (convert text → numbers)
# ============================================================
# STEP 6: ENCODE THE TARGET COLUMN
# ML models only understand numbers, not "acquired"/"closed"
# ============================================================

# Convert: "acquired" → 1 (success), "closed" → 0 (failure)
df['status'] = df['status'].map({'acquired': 1, 'closed': 0})

print(" Target column encoded:")
print(df['status'].value_counts())
print(f"\n1 = Acquired (Success) : {(df['status']==1).sum()} startups")
print(f"0 = Closed  (Failure)  : {(df['status']==0).sum()} startups")
```

```python
# Cell 9 — encode categorical text columns
# ============================================================
# STEP 7: ENCODE CATEGORICAL COLUMNS
# city, state_code, category_code are text — must be numbers
# ============================================================

le = LabelEncoder()

# Encode city
if 'city' in df.columns:
    df['city_encoded'] = le.fit_transform(df['city'].astype(str))
    df.drop('city', axis=1, inplace=True)
    print(" 'city' → 'city_encoded'")

# Encode state_code
if 'state_code' in df.columns:
    df['state_encoded'] = le.fit_transform(df['state_code'].astype(str))
    df.drop('state_code', axis=1, inplace=True)
    print(" 'state_code' → 'state_encoded'")

# Encode category_code
if 'category_code' in df.columns:
    df['category_encoded'] = le.fit_transform(df['category_code'].astype(str))
    df.drop('category_code', axis=1, inplace=True)
    print(" 'category_code' → 'category_encoded'")

print(f"\nDataset shape now: {df.shape}")
```

```python
#Cell 10 — feature engineering (infinity-safe version)
# ============================================================
# STEP 8: FEATURE ENGINEERING — FIXED (infinity-safe)
# ============================================================

# Feature 1: Funding efficiency (safe division)
df['funding_efficiency'] = (
    df['funding_total_usd'] / (df['funding_rounds'] + 1)
)

# Feature 2: Investor network strength (no division — safe)
df['investor_strength'] = (
    df['has_VC'] * 3 +
    df['has_angel'] * 2 +
    df['has_roundA'] +
    df['has_roundB'] +
    df['has_roundC'] +
    df['has_roundD']
)

# Feature 3: Milestone speed — use +1 to prevent division by zero
df['milestone_speed'] = (
    df['milestones'] / (df['age_first_milestone_year'].clip(lower=0.1) + 1)
)

# Feature 4: Network funding ratio — use log safely
df['network_funding_ratio'] = (
    df['relationships'] / (np.log1p(df['funding_total_usd'].clip(lower=1)) + 1)
)

# Feature 5: Top startup state flag
df['is_top_state'] = (
    (df['is_CA'] | df['is_NY'] | df['is_MA']).astype(int)
)

# Feature 6: Log of funding (clip to 1 so log(0) never happens)
df['log_funding'] = np.log1p(df['funding_total_usd'].clip(lower=0))

# ============================================================
# CRITICAL FIX: Replace inf / -inf / NaN created by any division
# ============================================================
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Fill any remaining NaN with column median (safe fallback)
for col in df.select_dtypes(include=[np.number]).columns:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].median(), inplace=True)

# Verify: no infinity, no NaN left
inf_count = np.isinf(df.select_dtypes(include=[np.number])).sum().sum()
nan_count = df.isnull().sum().sum()
print(f"Infinity values remaining : {inf_count}")
print(f"NaN values remaining      : {nan_count}")
print(f"Dataset shape             : {df.shape}")
print(f"\n New features created:")
for f in ['funding_efficiency', 'investor_strength', 'milestone_speed',
          'network_funding_ratio', 'is_top_state', 'log_funding']:
    print(f"  → {f}: mean={df[f].mean():.4f}, max={df[f].max():.4f}")
```

```python
#Cell 11 — remove outliers using IQR method
# ============================================================
# STEP 9: HANDLE OUTLIERS (IQR Method)
# ============================================================

outlier_cols = ['funding_total_usd', 'avg_participants', 'relationships']

print("Before outlier removal:", df.shape)

for col in outlier_cols:
    Q1 = df[col].quantile(0.25)   # 25th percentile
    Q3 = df[col].quantile(0.75)   # 75th percentile
    IQR = Q3 - Q1                   # Interquartile Range
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    before = df.shape[0]
    df = df[(df[col] >= lower) & (df[col] <= upper)]
    removed = before - df.shape[0]
    print(f" '{col}': removed {removed} outliers | range [{lower:.0f}, {upper:.0f}]")

print(f"\n After outlier removal: {df.shape}")
```

```python
#Cell 12 — scale features (normalize all numbers to same range)
# ============================================================
# STEP 10: FEATURE SCALING
# All features must be on the same scale for ML/ANN/CNN
# ============================================================

# Separate features (X) and target (y)
X = df.drop('status', axis=1)
y = df['status']

print(f"Features (X) shape: {X.shape}")
print(f"Target   (y) shape: {y.shape}")

# Apply StandardScaler: transforms each column to mean=0, std=1
# This is CRITICAL for ANN and CNN — they are sensitive to scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert back to DataFrame to keep column names
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print(f"\n Scaling complete!")
print(f"Before scaling - funding_total_usd mean: {X['funding_total_usd'].mean():.0f}")
print(f"After  scaling - funding_total_usd mean: {X_scaled['funding_total_usd'].mean():.4f}")
```
